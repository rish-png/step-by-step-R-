## Hierarchical Clustering
Divisive method using Euclidean distances.

<ins>Pre-requisite</ins>: package dplyr must be isntalled. To install, use `package.install('dplyr')` in R console.

```r
library(datasets)
library(dplyr)
```

```r
head(mtcars)
```

```r
cars <- mtcars[, c(1:4, 6:7, 9:11)] # Skipping column 5 and 8
head(cars)
```

```r
hclust(dist(cars))
```

```r
#using pipes from dplyr
hc <- cars %>% 
      dist %>%       # Compute distance/dissimilarity matrix
      hclust         # Compute hierarchical clusters
```

```r
plot(hc)
```

```r
plot(hc)
rect.hclust(hc, k=2, border='gray')
rect.hclust(hc, k=3, border='blue')
rect.hclust(hc, k=4, border='green')
rect.hclust(hc, k=5, border='red')
```
