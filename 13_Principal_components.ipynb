## Dimenstionality Reduction(Principal Components Analysis)

```r
library(datasets)
```

```r
head(mtcars)
```

```r
cars <- mtcars[, c(1:4, 6:7, 9:11)] # Skipping column 5 and 8
head(cars)
```

```r
pc <- prcomp(cars, center=TRUE, scale=TRUE)
```

```r
summary(pc)
```

```r
plot(pc)
```
